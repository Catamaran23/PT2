```python
#pretraitement
new_cols = ["ID", "Date", "Heure", "Description", "Adresse IP", "Nom d'hôte",
            "Adresse MAC", "Nom d'utilisateur", "ID transaction", "Résultat Q",
            "Heure essai", "ID corrélation", "DHCID", "ClasseFournisseur (Hex)",
            "ClasseFournisseur (ASCII)", "ClasseUtilisateur (Hex)", "ClasseUtilisateur (ASCII)",
            "InformationsAgentRelais", "ErreurEnregistrementDNS"]
dhcplog[new_cols] = dhcplog['Unnamed: 0'].str.split(',', expand=True)
```

```python
dhcplog=dhcplog.iloc[1:,0:]
dhcplog = dhcplog.drop(columns=['Unnamed: 0','Unnamed: 1'])
```

```python
dhcplog=dhcplog[["ID", "Date", "Heure","Adresse IP", "Nom d'hôte",
            "Adresse MAC", "Nom d'utilisateur", "ID transaction", "Résultat Q",
            "Heure essai", "ID corrélation", "DHCID", "ClasseFournisseur (Hex)",
            "ClasseFournisseur (ASCII)", "ClasseUtilisateur (Hex)", "ClasseUtilisateur (ASCII)",
            "InformationsAgentRelais", "ErreurEnregistrementDNS"]]
dhcplog
```

```python
new1=dhcplog.sort_values(by =["Adresse IP",'Date',"Heure"])
new1
```

```python
new1=new1.iloc[579:]
new1=new1.dropna()
new1.head(50)
```

```python
new1['Date'] = pd.to_datetime(new1['Date'], format='%m/%d/%y',errors='coerce')
new1['DateTime'] = new1['Date'].astype(str) + ' ' + new1['Heure']
new1['DateTime'] = pd.to_datetime(new1['DateTime'], errors='coerce')
new1['SessionDuration'] = new1.groupby(["Adresse IP","Date"])["DateTime"].diff()
new1.head(30)
```

```python
new1=new1[["ID", "Date", "Heure","Adresse IP", "Nom d'hôte",
            "Adresse MAC","DateTime","SessionDuration"]]
new1=new1.iloc[1:]
new1
```

```python
new1 = new1.dropna(subset=['SessionDuration'])
Q1 = new1['SessionDuration'].quantile(0.25)
Q3 = new1['SessionDuration'].quantile(0.75)
IQR = Q3 - Q1

seuil_bas = Q1 - 1.5 * IQR
seuil_haut = Q3 + 1.5 * IQR
```

```python
new1['Anomalies']=0
new1.loc[(new1['SessionDuration'] < seuil_bas) | (new1['SessionDuration'] > seuil_haut), 'Anomalies'] = 1
```
