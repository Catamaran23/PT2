```python
new1=dhcplog.sort_values(by =["Adresse IP",'Date',"Heure"])
new1
```

```python
new1=new1.iloc[579:]
new1=new1.dropna()
new1.head(50)
```

```python
new1['Date'] = pd.to_datetime(new1['Date'], format='%m/%d/%y',errors='coerce')
new1['DateTime'] = new1['Date'].astype(str) + ' ' + new1['Heure']
new1['DateTime'] = pd.to_datetime(new1['DateTime'], errors='coerce')
new1['SessionDuration'] = new1.groupby(["Adresse IP","Date"])["DateTime"].diff()
new1.head(30)
```

```python
new1=new1[["ID", "Date", "Heure","Adresse IP", "Nom d'hôte",
            "Adresse MAC","DateTime","SessionDuration"]]
new1=new1.iloc[1:]
new1
```

```python
new1 = new1.dropna(subset=['SessionDuration'])
Q1 = new1['SessionDuration'].quantile(0.25)
Q3 = new1['SessionDuration'].quantile(0.75)
IQR = Q3 - Q1

seuil_bas = Q1 - 1.5 * IQR
seuil_haut = Q3 + 1.5 * IQR
Outliers=new1[(new1['SessionDuration'] < seuil_bas) | (new1['SessionDuration'] > seuil_haut)]
```

```python
new1['Anomalies']=0
new1.loc[(new1['SessionDuration'] < seuil_bas) | (new1['SessionDuration'] > seuil_haut), 'Anomalies'] = 1
```

```python
newdataset=new1.drop(columns=["Adresse MAC","Nom d'hôte","ID"])
newdataset
```

```python
newdataset_modele=newdataset.drop(columns=["Date","Heure","DateTime"])
newdataset_modele['SessionDuration'] = newdataset_modele['SessionDuration'].dt.total_seconds()
newdataset_modele
```

```python
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import socket
import struct
```

```python
def ip_to_int(ip):
    return struct.unpack("!I", socket.inet_aton(ip))[0]
```

```python
newdataset_modele['IP_Int'] = newdataset_modele['Adresse IP'].apply(ip_to_int)
```

```python
X=newdataset_modele[["IP_Int","SessionDuration"]].values
X_outliers =Outliers[["Adresse IP","SessionDuration"]]
Y=newdataset_modele['Anomalies'].values
```
